```python
import numpy as np
import matplotlib.pyplot as plt

from keras.preprocessing import image
def prepare(img_path):
    img = load_img(img_path, target_size=(300,300))
    x = img_to_array(img)
    x = x/255
    return np.expand_dims(x, axis=0)
```
